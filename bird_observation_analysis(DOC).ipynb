```python
#Handling Missing Values
#Drop columns with too many null values
forest_data.dropna(axis=1, thresh=int(0.8*len(forest_data)), inplace=True)
grassland_data.dropna(axis=1, thresh=int(0.8*len(grassland_data)), inplace=True)

#Fill missing numerical values with median or mean where appropriate
forest_data['temperature'].fillna(forest_data['temperature'].median(), inplace=True)
grassland_data['temperature'].fillna(grassland_data['temperature'].median(), inplace=True)
#Median is used when the data is skewed, as it’s less sensitive to outliers.

#Drop rows where critical columns (like "observer") are missing
#Using inplace=True will modify the original DataFrame directly
forest_data.dropna(subset=['observer'], inplace=True) 
grassland_data.dropna(subset=['observer'], inplace=True)

#Short Note
#Dropping columns that have more than 20% missing data (thresh=0.8)
#Filling missing numerical values in the temperature column using the median, which is less affected by outliers
#Removing rows where critical information like the observer column is missing, as it's necessary for analysis

#Keyword Description
#dropna(): Removes missing (NaN) values from the dataset
  #axis=1: Drops columns with missing values
  #thresh: Sets a threshold for the minimum number of non-null values a column must have to be kept
  #subset: Specifies columns where missing values will be considered before dropping rows

#fillna()  : Replaces missing values with a specified value (e.g., median, mean)
  #median(): Computes the median (middle value) for the column and fills missing values with it 
    #The median is used to avoid the influence of outliers

#inplace=True: Modifies the original DataFrame without needing to assign it back to the variable
```

```python
#Reviewing - After cleaned version
#Display columns with missing values (if any)
missing_forest = forest_data.isnull().sum()
missing_grassland = grassland_data.isnull().sum()

#Filter out columns with no missing values for cleaner output
missing_forest = missing_forest[missing_forest > 0]
missing_grassland = missing_grassland[missing_grassland > 0]

print("\nMissing values in Forest data:\n", missing_forest)
print("\nMissing values in Grassland data:\n", missing_grassland)

#Short Note
#Code block checks for and displays only the columns with missing values in both forest and grassland datasets. 
  #Counts the number of missing values in each column
  #Filters out columns that have no missing values
  #Prints a cleaner summary showing only the columns that require attention for missing data handling

#Keyword Description
#isnull()  : Identifies missing (NaN) values in the DataFrame
#sum()     : Counts how many missing values are present in each column
#> operator: Filters to include only columns that have more than 0 missing values
#print()   : Displays the final filtered result of missing columns
```

```python
#Missing Values Handling:
#Distance: Filled missing values with the most frequent value (mode) since it's categorical.
#Acceptedtsn (Forest): One missing value, filled with 'Unknown'.
#Id_method & Taxoncode (Grassland): Filled missing values with 'Not Available' or dropped rows based on relevance.
```

```python
#For 'distance' (categorical data), fill missing values with the most frequent value
forest_data['distance'] = forest_data['distance'].fillna(forest_data['distance'].mode()[0])
grassland_data['distance'] = grassland_data['distance'].fillna(grassland_data['distance'].mode()[0])

#For 'acceptedtsn', cast to object and fill missing values with 'Unknown' for Forest data
forest_data['acceptedtsn'] = forest_data['acceptedtsn'].astype('object')
forest_data['acceptedtsn'] = forest_data['acceptedtsn'].fillna('Unknown')

#For 'id_method', cast to object and fill missing values with 'Not Specified' for Grassland data
grassland_data['id_method'] = grassland_data['id_method'].astype('object')
grassland_data['id_method'] = grassland_data['id_method'].fillna('Not Specified')

#For 'taxoncode', cast to object and fill missing values with 'Not Available' for Grassland data
grassland_data['taxoncode'] = grassland_data['taxoncode'].astype('object')
grassland_data['taxoncode'] = grassland_data['taxoncode'].fillna('Not Available')

#Short Note
#Handles missing values for categorical columns:
  #Fills missing values in 'distance' (categorical) with the most common value (mode()), which is a common approach for text categories
  #Converts acceptedtsn in the forest dataset to a string and fills blanks with "Unknown"
  #Converts id_method and taxoncode in the grassland dataset to strings and fills them with "Not Specified" and "Not Available" respectively
    
#Keyword Description
#fillna()        : Replaces missing values in a column with a specified value
#mode()          : Returns the most frequent value in a column (used for categorical data)
#astype('object'): Ensures the column is treated as categorical/text data (string/object type)
  #Categorical Data: Data that represents categories or groups (like colors, names, or types); [Type: dog, cat, bird]
  #Numerical Data  : Data that represents numbers you can measure or count (like age, height, or temperature)
```

```python
#Time of Day Analysis

#Observation Time: Study the Start_Time and End_Time to determine if specific time windows correlate with higher bird activity.

#Time of Day Analysis – Create Time Period Buckets

import pandas as pd
import matplotlib.pyplot as plt

# Load Data from Excel
forest_data = pd.read_excel(
    r"C:\Users\Bala Sowntharya\Documents\Project 2 - Bird Species Observation Analysis in Forest and Grassland Ecosystem\data_raw_excel files\Bird_Monitoring_Data_FOREST.XLSX"
)
grassland_data = pd.read_excel(
    r"C:\Users\Bala Sowntharya\Documents\Project 2 - Bird Species Observation Analysis in Forest and Grassland Ecosystem\data_raw_excel files\Bird_Monitoring_Data_GRASSLAND.XLSX"
)

# Strip any leading/trailing whitespace from column names
forest_data.columns = forest_data.columns.str.strip()

# Check if both 'Date' and 'Start_Time' columns exist
if 'Start_Time' in forest_data.columns and 'Date' in forest_data.columns:
    # Combine 'Date' and 'Start_Time' into full datetime
    forest_data['Start_Time'] = pd.to_datetime(
        forest_data['Date'].astype(str) + ' ' + forest_data['Start_Time'].astype(str),
        errors='coerce'
    )

    # Replace year 2022 with 2018
    forest_data['Start_Time'] = forest_data['Start_Time'].apply(
        lambda x: x.replace(year=2018) if pd.notna(x) and x.year == 2022 else x
    )

    # Print how many rows are in 2022 (should be 0) and 2018 (should be most)
    print("Year counts in 'Start_Time':")
    print(forest_data['Start_Time'].dt.year.value_counts())

    # Check for missing or invalid dates in 'Start_Time'
    invalid_dates = forest_data['Start_Time'].isna().sum()
    print(f"Invalid or missing 'Start_Time': {invalid_dates}")

    if invalid_dates == len(forest_data):
        print("All 'Start_Time' entries are invalid. No data to plot.")
    else:
        # Extract the hour from 'Start_Time' and create 'start_hour'
        forest_data['start_hour'] = forest_data['Start_Time'].dt.hour

        # Categorize time of day based on hour
        def map_time_of_day(hour):
            if pd.isna(hour):
                return 'Unknown'
            elif 5 <= hour < 12:
                return 'Morning'
            elif 12 <= hour < 17:
                return 'Afternoon'
            elif 17 <= hour < 21:
                return 'Evening'
            else:
                return 'Night'

        forest_data['time_of_day'] = forest_data['start_hour'].apply(map_time_of_day)

        # Clean data by removing 'Unknown' time_of_day entries
        forest_data_clean = forest_data[forest_data['time_of_day'] != 'Unknown']

        # Check how many valid rows remain after cleaning
        print(f"Rows after cleaning (valid time_of_day): {len(forest_data_clean)}")

        # Extract date from 'Start_Time'
        forest_data_clean['date'] = forest_data_clean['Start_Time'].dt.date

        # Group by date and time_of_day and count
        daily_time_counts = forest_data_clean.groupby(['date', 'time_of_day']).size().unstack(fill_value=0)

        print(daily_time_counts.head())

        if daily_time_counts.empty:
            print("No data available for plotting.")
        else:
            # Plotting the result
            ax = daily_time_counts.plot(kind='bar', stacked=True, figsize=(12, 8),
                                        color=['lightcoral', 'lightsalmon', 'lightgreen', 'lightblue'],
                                        edgecolor='black')

            plt.title('Time of Day Distribution per Day - Bird Observations (2018 Data)')
            plt.xlabel('Date')
            plt.ylabel('Number of Observations')
            plt.xticks(rotation=45)
            plt.legend(title='Time of Day', labels=['Morning', 'Afternoon', 'Evening', 'Night'],
                       bbox_to_anchor=(1.05, 1), loc='upper left')

            # Add data labels
            for p in ax.patches:
                height = p.get_height()
                if height > 0:
                    ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., p.get_y() + height / 2),
                                ha='center', va='center', fontsize=9, color='black')

            plt.tight_layout()
            plt.show()

else:
    print("'Start_Time' or 'Date' column is missing.")

#Short Note: Short Note: Combined date and time to create full timestamps, categorized bird observations into time-of-day buckets (Morning, Afternoon, Evening, Night), removed invalid entries, and plotted a stacked bar chart to show daily observation patterns across different times of day.

#Keyword/Function	                   Short Explanation
#fillna()	      -   Replaces missing (NaN) values with a specified value
#pd.to_datetime()  -	 Converts a column to datetime format; invalid values become NaT (if errors='coerce')
#.dt.hour	      -   Extracts the hour from a datetime column
#.dt.date	      -   Extracts just the date (YYYY-MM-DD) from a datetime column
#.apply()	      -   Applies a custom function to each row or element in a column
#.groupby()	      -   Groups the data by one or more columns for aggregation
#.size()	          -   Returns the size (number of rows) per group in groupby
#.unstack()	      -   Converts a grouped DataFrame into a wide format
#.isna() or .isnull()   - Checks for missing (NaN) values
#.notna() or .notnull() - Checks for non-missing (valid) values
#inplace=True	  -   Applies the operation directly to the DataFrame (no need to assign it again)
#errors='coerce' (in to_datetime)	Invalid parsing will be set to NaT (Not a Time), avoiding crashes


#.plot()	     -  Creates basic visualizations (line, bar, etc.) from DataFrames
#kind='bar'	     -  Specifies the type of plot (bar chart in this case)
#stacked=True    -	Stacks values in bar plots for grouped columns
#figsize=(x, y)  - Sets the size of the plot (width, height in inches)
#plt.xticks(rotation=45)	  - Rotates x-axis labels for better readability
#.annotate() (from matplotlib)-	Adds text labels to the plot (e.g., value labels on bars)

#Purpose: Analyze bird observation activity by time of day using Start_Time data from forest observations

#Step-by-Step Summary:
#Load Excel Data: Read bird monitoring data from forest and grassland Excel files
#Clean Columns:   Strip spaces from column headers to avoid key errors
#Create Datetime: Combine Date and Start_Time into a single datetime column
#Fix Year:        Replace all 2022 entries with 2018 to maintain consistency
#Extract Hour & Categorize: Use the hour from Start_Time to classify observations into:
     #Morning (5–12), Afternoon (12–17), Evening (17–21), Night (Others)
#Clean Data:     Remove rows with missing or invalid Start_Time
#Group & Count: Group data by date and time of day to count observations
#Visualize: Plot a stacked bar chart showing daily bird activity segmented by time of day, with data labels and legend
```

```python
#Categorizing Observations Based on Evening Time Slot

import pandas as pd

#Sample DataFrame (replace this with your actual data)
data = {
    'Start_Time': ['06:00:00', '07:30:00', '09:00:00', '17:00:00', '19:30:00', '21:00:00']
}
forest_data = pd.DataFrame(data)

#Convert 'Start_Time' to a datetime object if it's just time (you may need to adjust for your actual dataset)
default_date = '2023-01-01'  # Default date to combine with time
forest_data['Start_Time'] = pd.to_datetime(default_date + ' ' + forest_data['Start_Time'].astype(str), errors='coerce')

#Create a new 'start_hour' column
forest_data['start_hour'] = forest_data['Start_Time'].dt.hour

#Create a new column to flag evening times (for example, consider evening as 6 PM (18:00) to 9 PM (21:00))
def categorize_evening(row):
    if 18 <= row['start_hour'] <= 21:
        return 'Evening'
    return 'Not Evening'

forest_data['time_of_day'] = forest_data.apply(categorize_evening, axis=1)

#Check the data
print(forest_data[['Start_Time', 'start_hour', 'time_of_day']].head())

#Short Note: Cconverted the Start_Time column to a datetime format by appending a default date, 
#Extracted the hour from each time entry, and then classified each row as "Evening" if the observation time was between 6 PM and 9 PM, inclusive. 
#This helps identify observations recorded during evening hours


#Keyword Description
#pd.to_datetime()	   Converts string or object data to datetime format for easier time-based operations
#astype(str)	       Converts a column to string type, useful for formatting and concatenation
#dt.hour	           Extracts the hour part from a datetime object
#apply()	           Applies a function row-wise to the DataFrame
#categorize_evening()  Custom function to flag times between 18:00–21:00 as "Evening"
#'time_of_day'	       New column indicating whether the observation occurred in the evening or not
```

```python
#Using Geopandas for Geospatial Data (with Shapefiles)
#Will plot points on a world map, where the size of the point correlates with the observation count
import geopandas as gpd                 #For spatial data manipulation
import pandas as pd                     #For data handling
import matplotlib.pyplot as plt         #For plotting
from shapely.geometry import Point      #For creating geometries
import matplotlib.patches as mpatches   #For custom plot patches (legends, shapes)


# Load world boundaries (download once and reuse)
world = gpd.read_file("https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/geojson/ne_110m_admin_0_countries.geojson")

# Example site data with lat/lon and observation counts
data = pd.DataFrame({
    'site_name': ['ANTI 1', 'ANTI 2', 'SITE 3'],
    'latitude': [12.9716, 13.0827, 13.2891],
    'longitude': [77.5946, 80.2707, 78.9629],
    'count': [182, 151, 200]
})

# Convert to GeoDataFrame
geometry = [Point(xy) for xy in zip(data['longitude'], data['latitude'])]
gdf = gpd.GeoDataFrame(data, geometry=geometry, crs="EPSG:4326")

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))
world.plot(ax=ax, color='lightgray', edgecolor='white')

# Plot latitude in one color and longitude in another color
gdf.plot(ax=ax, color='red', markersize=gdf['count'], alpha=0.6, label="Latitude")
gdf.plot(ax=ax, color='skyblue', markersize=gdf['count'], alpha=0.6, marker='o', label="Longitude")

# Add a custom legend
latitude_patch = mpatches.Patch(color='red', label='Latitude')
longitude_patch = mpatches.Patch(color='skyblue', label='Longitude')
plt.legend(handles=[latitude_patch, longitude_patch], loc='upper right')

plt.title("Geographical Distribution of Observations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.tight_layout()
plt.show()

#Keyword                     Explanation
#geopandas	         Library for working with geospatial data (maps + data)
#gpd.read_file()	 Loads geographic shapefiles/geoJSON (world boundaries here)
#Point()	         Converts latitude & longitude into geometric points
#GeoDataFrame	     A special DataFrame that stores geometry with data
#EPSG:4326	         Coordinate Reference System (standard for lat/lon)
#plot()	             Plots the world map and the points (site observations)
#markersize	         Controls the size of each point based on observation count
#alpha	             Sets the transparency of points (0: transparent, 1: opaque)
#tight_layout()	     Adjusts layout to avoid label cutoff
#mpatches (matplotlib.patches)

#mpatches is a module from matplotlib used to create shapes (like rectangles, circles) for customizing plots, adding annotations, and creating custom legends

#Short Description
#Used GeoPandas to plot bird observation sites on a world map, where each point represents a site and its size reflects the observation count
#We converted latitude and longitude into geometry points, created a GeoDataFrame, and overlaid it on a world basemap using matplotlib

#Data Taken from open-source platform
   #The data was taken from an open-source platform like eBird or GBIF, which provides bird observation records with location and count. 
   #We filtered and exported the relevant data into Excel, then used it to visualize high-activity zones using maps and heatmaps.
```

```python
#Reference
#Clean and convert the wind column

#Convert wind descriptions to numerical values, such as wind speed ranges
def categorize_wind(wind_description):
    if 'Calm' in wind_description:
        return 0  #Calm winds
    elif 'Light' in wind_description:
        return 1  #Light winds
    elif 'Moderate' in wind_description:
        return 2  #Moderate winds
    elif 'Strong' in wind_description:
        return 3  #Strong winds
    elif 'Gale' in wind_description:
        return 4  #Gale force winds
    else:
        return 0  #Default to Calm for any unrecognized description

#Apply the function to the wind column
forest_data['Wind_numeric'] = forest_data['Wind'].apply(categorize_wind)

#Now, check the cleaned 'wind_numeric' column
print(forest_data['Wind_numeric'].unique())

#Keyword Explanation

#def             : Defines a function in Python
#if / elif / else: Conditional statements to check multiple conditions
#in              : Checks if a substring exists in a string
#return          : Sends back a value from a function
#apply()         : Applies a function to each element in a DataFrame column
#unique()        : Returns the unique values in a column

#Short Description
#It's  categorizes wind descriptions into numerical values (e.g., Calm, Light, Moderate) and applies this categorization to a new column (Wind_numeric) in the forest_data DataFrame. 
#It then displays the unique values in the new column.
```
